```python
import numpy as np
import pandas as pd
import astropy
from astroquery.simbad import Simbad
```

```python
# Not sure what the best way to read in the binary names from simbad_eclipsing_binaries_20000_stars.txt is - it's pretty weirdly formatted without a set delimiter
# Let me first try the pandas read_table() function
data = pd.read_table('simbad_eclipsing_binaries_20000_stars.txt',sep=None,header=None,names=['#','identifier','typ','coord1','Mag U','Mag B','Mag V','Mag R',
                                                                                           'Mag I','spec. type','#bib','#not'],engine='python',skiprows=9,nrows=10)
```

```python
data
```

```python
# OK, I could have predicted that would go badly. I think the best way to do this (other than potentially using astropy.io.ascii.read(), which I'll do as a last
# resort), is just to use the basic Python with open(): file-reading mechanism. That way I'll have a little more control over how each line is read. Let's try:
names = []
with open('simbad_eclipsing_binaries_20000_stars.txt') as f:
    # There are 9 junk (non-data) lines at the beginning of the file and 3 junk lines at the end
    for i, line in enumerate(f):
        if i < 9:
            continue
        string = line
        ind1 = string.find('|')+1
        if ind1 == 0: # If this is the case, then we're at the end of the file in those last 3 lines (can skip)
            break
        name = string[(ind1):(string[5:].find('  ')+5)] # Based on the structure of the data file, this should give us the name of each star ('identifier')
        names.append(name)
```

```python
# OK, that actually worked great, so now I have a list of eclipsing binary names
# Now how to get the TIC IDs with astroquery?
# Looking at the following link: https://astroquery.readthedocs.io/en/latest/simbad/simbad.html
# Apparently you can't submit more than 5-10 queries per second to SIMBAD without being blacklisted, which is an issue for what I was planning on doing (looping 
# through all the targets and submitting one query for each). However, apparently you can use vectorized queries by passing a list of object names to
# query_objects, and SIMBAD treats it as a single query
# There is a section on the above web page called "Query object identifiers" and another called "Vectorized queries" - these 2 combined are what I want
result_table = Simbad.query_objectids("V* V662 And")
print(type(result_table))
```

```python
print(result_table)
```

```python
# Let's test out how this might work for one binary
ticIDs = [] # Empty list to hold the TIC ID of our binary
for ID in result_table['ID']: # Loop through every alternative identifier for our object
    var = ID.find('TIC')
    if var != -1: # If 'TIC' appears in the current identifier - i.e., it's the TIC ID
        ticID = ID[ID.find(' ') + 1:] # Get the actual numerical TIC ID
        ticIDs.append(ticID) # Add this TIC ID to the list of TIC IDs
```

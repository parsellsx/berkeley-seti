```python
# OK, I could have predicted that would go badly. I think the best way to do this (other than potentially using astropy.io.ascii.read(), which I'll do as a last
# resort), is just to use the basic Python with open(): file-reading mechanism. That way I'll have a little more control over how each line is read. Let's try:
names = []
with open('simbad_eclipsing_binaries_20000_stars.txt') as f:
    # There are 9 junk (non-data) lines at the beginning of the file and 3 junk lines at the end
    for i, line in enumerate(f):
        if i < 9:
            continue
        string = line
        ind1 = string.find('|')+1
        if ind1 == 0: # If this is the case, then we're at the end of the file in those last 3 lines (can skip)
            break
        name = string[(ind1):(string[5:].find('  ')+5)] # Based on the structure of the data file, this should give us the name of each star ('identifier')
        names.append(name)
```
